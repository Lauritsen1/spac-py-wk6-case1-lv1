```python
import string

import pandas as pd
```

```python
df = pd.read_csv('../data/raw/tøjfirma_sales_data.csv')
df.head(10)
```

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

```python
cols = ['Product', 'Category', 'Size']
df[cols] = df[cols].apply(lambda x: x.str.strip(string.punctuation))
```

```python
df['Sale Price'] = df['Sale Price'].round(2)
```

```python
df['Sale Date'] = pd.to_datetime(df['Sale Date'])
```

```python
df.to_csv('../data/processed/tøjfirma_sales_data.csv', index=False)
```
